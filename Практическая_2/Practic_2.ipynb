```python
contries = ['United States', 'Great Britain',
         'China', 'Russia', "Germany"]

gold_medal = [46, 27, 26, 19, 17]
explode = (0.1, 0, 0, 0, 0)

plt.pie(gold_medal, explode = explode, labels = contries,
        autopct = '%1.1f%%',shadow = True,
        startangle = 90,
        wedgeprops = {"edgecolor":"black",
                    'linewidth': 2,
                    'antialiased': True})

# Equal aspect ratio ensures
# that pie is drawn as a circle.
plt.axis('equal')

plt.show()
```

```python
import pandas as pd
```

```python
df = pd.read_csv('global_temps.csv')
df.head()
```

```python
x = df['Year'].values
y1 = df.iloc[:, 1:13].mean(axis = 1).values
x2 = df.columns.values[1:13]
y2 = df[df["Year"] == 2022].iloc[:, 1:13].values[0]
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.15, 0.55, 0.5, 0.3])
axes1.plot(x, y1)
axes2.plot(x2, y2)
axes1.set_title('Средняя температура по годам')
axes2.set_title('Температура в течении 2022 г.')
```

Узнаем сколько лет средняя температура была меньше нуля и сколько лет больше

```python
x = ['меньше нуля','больше нуля']
x1 = [0, 1]
t = pd.DataFrame({'mean_temp':df.iloc[:, 1:13].mean(axis = 1)})
y = [t[t.mean_temp <= 0].count().values[0], t[t.mean_temp > 0].count().values[0]]
plt.title('Среднегодовая температура')
plt.bar(x1, y)
plt.xticks(x1, x)


plt.show()
```
